```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
filename = "pgaTourData.csv"

df = pd.read_csv(filename)

print("SHAPE")
print(df.shape)
print("\nINFO")
df.info()
print("\n DESCRIBE")
print(df.describe())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10,6))
sns.regplot(data=df, x="Avg Distance", y="Fairway Percentage")

plt.title('Pricksäkerheten blir mindre desto längre man slår med drivern', fontsize=14)
plt.xlabel('Average Driving Distance (Yards)', fontsize=12)
plt.ylabel('Fairway Percentage (%)', fontsize=12)

plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 6))

sns.lineplot(x="Year", y="Avg Distance",
             data=df, ax=ax1, color="forestgreen")
ax1.set_title("Trend: Driving Distance ökar med åren")
sns.lineplot(x="Year", y="Fairway Percentage",
             data=df, ax=ax2, color="darkred")
ax2.set_title ("Trend: Fairway Percentage minskar med åren")

plt.tight_layout()
plt.show()
```

En av den mest ställda frågan inom golf är om man ska lägga majoriteten av tiden på att slå längre och rakare utslag eller om det är precisionen med järnklubborna in mot green som sänker din score mest. Detta diagram har jag skapat för att ta reda på det baserat på vad vi ser på touren.

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(16, 6))

sns.regplot(data=df, x="SG:OTT", y="Average Score", ax=ax1, color="green")
ax1.set_title("Utslag vs. Score")

sns.regplot(data=df, x="SG:APR", y="Average Score", ax=ax2, color="blue")
ax2.set_title("Inspel vs. Score")

plt.show()
```

Svårt att se skillnaden på diagrammet då det är väldigt tight.

```python
print(f"Korrelation Utslag/Score: {df['SG:OTT'].corr(df['Average Score']):.4f}")
print(f"Korrelation Inspel/Score: {df['SG:APR'].corr(df['Average Score']):.4f}")

## Vi kan se att inspelet är den viktigaste nycklen till ett lägre score.
```

Hur viktigt är det egentligen att man ligger i fairway då? Vi har sett en minskning av fairway % med åren men hur mycket påverkar det vinnarna? Detta vill jag ta reda på med diagrammet under.

```python
import seaborn as sns
import matplotlib.pyplot as plt

df['Wins'] = df['Wins'].fillna(0)
df['Is_Winner'] = df['Wins'] > 0

plt.figure(figsize=(8, 6))

sns.barplot(data=df, x='Is_Winner', y='Fairway Percentage', palette='Greens')

plt.title('Träffar vinnare fler fairways än icke vinnare?')
plt.ylabel('Fairway Percentage (%)')
plt.xlabel('Har vunnit minst en gång')

plt.ylim(50, 70) 

plt.show()
```
