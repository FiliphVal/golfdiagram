```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

###  Inledning

I detta projekt har jag valt att analysera golfstatistik från pga touren mellan åren 2010–2018, från Kaggle. Datan är på säsongsbasis, där varje rad representerar en spelares samlade genomsnitt under en hel säsong.

Syftet är att undersöka vad som är viktigast för att vinna matcher och även sänka sin score via dessa frågor:

1. **Slaglängd vs. Pricksäkerhet:** Offrar man precision från tee ju längre man slår med drivern?
2. **Utslag vs. Inspel:** Vilket moment har starkast koppling till en spelares genomsnittliga score? Spelet från tee eller precisionen på inspelen mot green?
3. **Vinnarnas precision:** Krävs det att man är tourens mest pricksäkra spelare i fairwayträffar för att vinna turneringar, eller spelar det mindre roll?

```python
filename = "pgaTourData.csv"

df = pd.read_csv(filename)

print("SHAPE")
print(df.shape)
print("\nINFO")
df.info()
print("\n DESCRIBE")
print(df.describe())
```

### Inläsning och inspektion av data

Datasetet består av 2 312 rader och 18 kolumner. Med .info() ser vi att det saknas data för 634 rader i några kolumner, bland annat för Fairway Percentage och Strokes Gained.

Detta är vanligt inom golf på denna nivå och beror mest på dessa två faktorer:
* **Krav på antal rundor:** PGA touren kräver att en spelare genomför ett minsta antal tävlingsrundor under en säsong för att tas med i slutstatistiken.
* **Tekniska begränsningar:** Strokes Gained statistiken bygger på ShotLink systemet. Om en spelare har kört turneringar på banor som saknar detta system registreras ingen data för rundan.

```python
df['Wins'] = df['Wins'].fillna(0)
df['Top 10'] = df['Top 10'].fillna(0)

df = df.dropna(subset=['Rounds', 'Fairway Percentage', 'SG:OTT', 'Points'])

df.info()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

snitt_langd = df['Avg Distance'].mean()
snitt_fairway = df['Fairway Percentage'].mean()

plt.figure(figsize=(10,6))
sns.regplot(data=df, x="Avg Distance", y="Fairway Percentage", 
            scatter_kws={"color": "purple", "alpha": 0.3}, line_kws={"color": "red", "alpha": 0.8})

plt.axvline(snitt_langd, color='black', linestyle='--', label=f'Snittlängd på touren ({snitt_langd:.2f} yards)')
plt.axhline(snitt_fairway, color='black', linestyle=':', label=f'Snitt fairway ({snitt_fairway:.1f}%)')


plt.title('Pricksäkerheten blir mindre desto längre man slår med drivern', fontsize=14)
plt.xlabel('Genomsnittling längd på slag från tee (Yards)', fontsize=12)
plt.ylabel('Träffade fairways (%)', fontsize=12)
plt.grid(True, alpha=0.3)
plt.legend()
plt.show()
```

Jag valde ett spridningsdiagram med en röd trendlinje för att se om det finns ett samband mellan slaglängd och pricksäkerhet. Diagrammet visar tydligt att ju längre spelarna slår i genomsnitt, desto färre fairways träffar de  och referenslinjerna visar att de som slår allra längst nästan alltid ligger under tourens snitt i precision.

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 6))

sns.lineplot(x="Year", y="Avg Distance",
             data=df, ax=ax1, color="forestgreen")
ax1.set_title("Den genomsnittliga slaglängden ökar stadigt varje år")
ax1.set_xlabel('År')
ax1.set_ylabel('Slaglängd från tee (yards)')
ax1.grid(True, alpha=0.3)
sns.lineplot(x="Year", y="Fairway Percentage",
             data=df, ax=ax2, color="darkred")
ax2.set_title ("I samband med högre slaglängd minskar precisionen")
ax2.set_xlabel('År')
ax2.set_ylabel('Träffade fairways (%)')
ax2.grid(True, alpha=0.3)

fig.suptitle('Utvecklingen på touren: Distans prioriteras men på bekostnad av precision')

plt.tight_layout()
plt.show()
```

Här la jag två linjediagram bredvid varandra för att se hur sporten har utvecklats över tid mellan 2010 och 2018. Det är intressant eftersom vi ser att slaglängden ökar stadigt varje år samtidigt som precisionen åker nedåt, vilket bevisar att spelarna medvetet väljer att offra fairways för att tjäna meter.

```python
import matplotlib.pyplot as plt
import seaborn as sns

corr_ott = df['SG:OTT'].corr(df['Average Score'])
corr_apr = df['SG:APR'].corr(df['Average Score'])

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(16, 6))

sns.regplot(data=df, x="SG:OTT", y="Average Score", ax=ax1, scatter_kws={"color": "purple", "alpha": 0.3}, line_kws={"color": "red"})
ax1.set_title(f"Bättre spel från tee sänker scoren (Korrelation: {corr_ott:.2f})")
ax1.set_xlabel("Strokes Gained: Off the Tee (Slag sparade per runda)")
ax1.set_ylabel("Genomsnittlig Score (Slag per runda)")
ax1.grid(True, alpha=0.3)


sns.regplot(data=df, x="SG:APR", y="Average Score", ax=ax2, scatter_kws={"color": "purple", "alpha": 0.3}, line_kws={"color": "green"})
ax2.set_title(f"Bättre precision från inspel sänker scoren (Korrelation: {corr_apr:.2f})")
ax2.set_xlabel("Strokes Gained: Approach (Slag sparade per runda)")
ax2.set_ylabel("Genomsnittlig Score (Slag per runda)")
ax2.grid(True, alpha=0.3)

fig.suptitle("Precisionen på inspelen väger tyngre än spelet från tee", fontsize=14, fontweight='bold')

plt.tight_layout()
plt.show()
```

Jag placerade två regplots bredvid varandra för att jämföra om utslagen eller inspelen styr spelarnas totala score mest, och la in korrlationen. Resultatet är väldigt intressant då trendlinjen för inspelen (Approach) är lite brantare, vilket visar att precisionen på inspelen är viktigast för en låg score.

```python
df['Is_Winner'] = df['Wins'] > 0

tour_snitt = df['Fairway Percentage'].mean()

plt.figure(figsize=(8, 6))

sns.boxplot(data=df, x='Is_Winner', y='Fairway Percentage', color='lightgreen')

plt.axhline(tour_snitt, color='red', linestyle='--', label=f'Tour-snitt ({tour_snitt:.1f}%)')

plt.title('Vinnare har inte högre precision från tee än övriga spelare')
plt.ylabel('Träffade fairways (%)')
plt.xlabel('Har vunnit minst en turnering')
plt.legend()

plt.ylim(50, 70)

plt.show()
```

Istället för ett stapeldiagram valde jag boxplot  för att visa hela spridningen på precisionen och trunkerade y axeln mellan 50% och 70% för att kunna se skillnaderna. Det intressanta här är att vinnarna och de vinstlösa spelarna ligger rätt jämnt runt snitt värdet, vilket bevisar att man inte måste vara bäst på att hitta fairway för att plocka hem vinster.

## Avslutning

Min undersökning visar att man medvetet väljer distans före precision. Att slå längre leder till färre träffade fairways, men i och med att precisonen på inspelen är viktigaste så är det logiskt. Detta bekräftas av att turneringarnas vinnare i snitt inte har bättre precision från tee än övriga spelare.

Datasetet har dock begränsningar som missar viktiga grejer:

* **Det mentala spelet:** Siffrorna visar bara slutresultat, men nerver och press kan påverka detta.
* **Formtoppar döljs:** Datan visar årsbetyg. Vi ser inte om en vinnare varit jämn hela säsongen eller bara haft sin bästa form under just vinstveckan.
* **Väder och banor:** Banor är olika svåra och vädret skiftar dagligen, men allt slås ihop till ett genomsnitt.
