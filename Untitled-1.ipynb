```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
filename = "pgaTourData.csv"

df = pd.read_csv(filename)

print("SHAPE")
print(df.shape)
print("\nINFO")
df.info()
print("\n DESCRIBE")
print(df.describe())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10,6))
sns.regplot(data=df, x="Avg Distance", y="Fairway Percentage")

plt.title('Pricksäkerheten blir mindre desto längre man slår med drivern', fontsize=14)
plt.xlabel('Average Driving Distance (Yards)', fontsize=12)
plt.ylabel('Fairway Percentage (%)', fontsize=12)

plt.show()
```
