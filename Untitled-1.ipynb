```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```


### Inledning

Detta projekt är en statistisk och visuell utforskning av professionell golfstatistik från PGA  touren under perioden 2010 till 2018. Datamängden är hämtad från Kaggle (PGA Tour Data 2010–2018) och innehåller detaljerad information om bland annat spelares prestationer på banan, inklusive slaglängd, pricksäkerhet, Strokes Gained-statistik.

### Vad är det för data?
Varje rad i datasetet representerar en specifik spelares samlade statistik. Det betyder att vi inte tittar på enskilda slag eller enstaka rundor, utan på stabila trender och genomsnitt över tid för världens bästa golfspelare.

### Syfte med detta projekt
Inom golf diskuteras det ständigt vad som faktiskt krävs för att vinna och sänka sin score. Är det din driver distance från tee, eller är det precisionen i inspelet mot green? Med hjälp av "Detektivens metod" och explorativ dataanalys är syftet med denna notebook att besvara följande frågor:

1. **Slaglängd vs. Pricksäkerhet:** Hur påverkas spelares förmåga att träffa fairway när de slår bollen längre med drivern? Är det lönt att offra precision för att vinna distans?
2. **Nyckeln till lägre score:** Vilket spelaspekt har starkast korrelation med en spelares genomsnittliga score? Är det de långa spelet från utslag eller precisionen på inspelet ?
3. **Är det viktigt att träffa fairway?:** Träffar de spelare som faktiskt vinner turneringar fler fairways i genomsnitt än de som inte vinner, eller spelar det mindre roll bland de bästa golfarna?

Genom att använda matplotlib och seaborn ska vi gå från rådata till riktig visuell statistik.

```python
filename = "pgaTourData.csv"

df = pd.read_csv(filename)

print("SHAPE")
print(df.shape)
print("\nINFO")
df.info()
print("\n DESCRIBE")
print(df.describe())
```

#### 2. Inläsning av datan

Efter att ha laddat in datasetet och kört inspektion så kan vi se att datasetet består av 2 312 rader och 18 kolumner.

#### Analys av saknade värden
När vi kör .info() ser vi att det finns saknad data. Medan Player Name och Year har fullständiga data (2 312 rader), så har kolumner som Rounds, Fairway Percentage och flera Strokes Gained-kolumner endast 1 678 registrerade värden. Det saknas alltså data för 634 rader.

**Varför saknas dessa värden?**
Inom elitgolf beror detta bortfall oftast på två specifika, sportsliga faktorer snarare än tekniska fel i databasen:
1. **Minsta antal rundor:** För att en spelare ska tas med i den officiella slutstatistiken för en säsong på PGA-touren krävs det vanligtvis att man har spelat ett visst antal tävlingsrundor (ofta runt 40–50 rundor).
2. **Tekniska begränsningar:** Strokes Gained systemet bygger på ShotLink teknologi. Om en spelare har spelat turneringar utanför de banor som är utrustade med detta system under säsongen, kan viss data saknas.

Dessa tomma värden betyder att vi behöver göra en datatvätt i nästa steg för att inte få missvisande statistik.

```python
df['Wins'] = df['Wins'].fillna(0)
df['Top 10'] = df['Top 10'].fillna(0)

df = df.dropna(subset=['Rounds', 'Fairway Percentage', 'SG:OTT', 'Points'])

print("\n Finns det några nan kvar?")
df.info()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

snitt_langd = df['Avg Distance'].mean()
snitt_fairway = df['Fairway Percentage'].mean()

plt.figure(figsize=(10,6))
sns.regplot(data=df, x="Avg Distance", y="Fairway Percentage", 
            scatter_kws={"color": "purple", "alpha": 0.3}, line_kws={"color": "red", "alpha": 0.8})

plt.axvline(snitt_langd, color='black', linestyle='--', label=f'Snittlängd på touren ({snitt_langd:.2f} yards)')
plt.axhline(snitt_fairway, color='black', linestyle=':', label=f'Snitt fairway ({snitt_fairway:.1f}%)')


plt.title('Pricksäkerheten blir mindre desto längre man slår med drivern', fontsize=14)
plt.xlabel('Genomsnittling längd på slag från tee (Yards)', fontsize=12)
plt.ylabel('Träffade fairways (%)', fontsize=12)
plt.grid(True, alpha=0.3)
plt.legend()
plt.show()
```

I min första frågeställning ville jag undersöka om man offrar pricksäkerheten för att slå längre. 

Jag valde att göra ett scatterplot och lägga på en trendlinje. Det är det absolut enklaste och bästa sättet när man vill kasta ut massor av spelare i en graf och se om prickarna rör sig i en viss riktning. För att göra diagrammet ännu tydligare la jag in ett genomsnitt i mitten med två svarta referenslinjer en vertikal för snittlängden och en horisontell för snittprecisionen.

Trendlinjen lutar tydligt nedåt, vilket bekräftar att man offrar pricksäkerheten. De två svarta linjerna skapar ett intressant kors i diagrammet. Om vi ser på spelarna som befinner sig till höger om den vertikala snittlinjen dvs. de som slår längre än genomsnittet så kan vi se att den absoluta majoriteten av dem hamnar under den horisontella linjen för snittprecision. Detta ger ett bevis för att extrem distans gör att man offrar precisionen i touren.

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 6))

sns.lineplot(x="Year", y="Avg Distance",
             data=df, ax=ax1, color="forestgreen")
ax1.set_title("Den genomsnittliga slaglängden ökar stadigt varje år")
ax1.set_xlabel('År')
ax1.set_ylabel('Slaglängd från tee (yards)')
ax1.grid(True, alpha=0.3)
sns.lineplot(x="Year", y="Fairway Percentage",
             data=df, ax=ax2, color="darkred")
ax2.set_title ("I samband med högre slaglängd minskar precisionen")
ax2.set_xlabel('År')
ax2.set_ylabel('Träffade fairways (%)')
ax2.grid(True, alpha=0.3)

fig.suptitle('Utvecklingen på touren: Distans prioriteras men på bekostnad av precision')

plt.tight_layout()
plt.show()
```

Efter att ha sett att det finns ett tydligt samband mellan att slå långt och att missa fairway, ville jag undersöka om längden har förändrats. Har spelarna börjat slå längre med åren?

Jag valde att lägga två linjediagram (lineplots) bredvid varandra i en och samma figur. Linjediagram valde jag för att visa en utveckling/en trend över tid (i det här fallet mellan 2010 och 2018).

Graferna pratar för sig själva man ser dem bredvid varandra. Till vänster ser vi hur den genomsnittliga slaglängden pekar spikrakt uppåt efter 2013 för varje säsong. Till höger ser vi hur andelen träffade fairways dyker neråt under exakt samma tidsperiod. 

Detta visar en tydlig strategisk förändring. Spelarna och tränarna har uppenbarligen kommit fram till att det är mer värt att tjäna meter från tee och slå kortare inspel, även om det betyder att de oftare får spela bollen från ett svårare läge (ruffen).

Detta diagram förvånade mig väldigt mycket för jag visste inte att slaglängden hade utvecklats så mycket med åren, det får mig att vilja ta reda på hur utvecklingen hade sett ut fram tills nu (2026).

```python
import matplotlib.pyplot as plt
import seaborn as sns

corr_ott = df['SG:OTT'].corr(df['Average Score'])
corr_apr = df['SG:APR'].corr(df['Average Score'])

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(16, 6))

sns.regplot(data=df, x="SG:OTT", y="Average Score", ax=ax1, scatter_kws={"color": "purple", "alpha": 0.3}, line_kws={"color": "red"})
ax1.set_title(f"Bättre spel från tee sänker scoren (Korrelation: {corr_ott:.2f})")
ax1.set_xlabel("Strokes Gained: Off the Tee (Slag sparade per runda)")
ax1.set_ylabel("Genomsnittlig Score (Slag per runda)")
ax1.grid(True, alpha=0.3)


sns.regplot(data=df, x="SG:APR", y="Average Score", ax=ax2, scatter_kws={"color": "purple", "alpha": 0.3}, line_kws={"color": "green"})
ax2.set_title(f"Bättre precision från inspel sänker scoren (Korrelation: {corr_apr:.2f})")
ax2.set_xlabel("Strokes Gained: Approach (Slag sparade per runda)")
ax2.set_ylabel("Genomsnittlig Score (Slag per runda)")
ax2.grid(True, alpha=0.3)

fig.suptitle("Precisionen på inspelen väger tyngre än spelet från tee", fontsize=14, fontweight='bold')

plt.tight_layout()
plt.show()
```

Efter att ha sett statistik på hur touren utvecklats över tid ville jag gräva djupare i vad det viktigaste är för ett lägre score.

Jag valde att lägga två spridningsdiagram bredvid varandra för att direkt kunna jämföra trendlinjerna mot samma y axel (Genomsnittlig score). Linjerna är hyfsat lika och därför använde jag metoden .corr för att se vilken som faktist är den viktigaste.

Båda trendlinjerna lutar nedåt, vilket är helt logiskt. Ju fler slag du sparar (Strokes Gained går plus), desto lägre och bättre blir din genomsnittliga score. 

Det intressanta här är dock att jämföra lutningen och korrelationen:
* Utslagen (vänster): Har en tydlig koppling till en lägre score.
* Inspelen (höger): Har en ännu brantare trendlinje och ett starkare korrelationsvärde. 

Detta visar att även om det är viktigt att slå bra från tee, så är precisionen på inspelen mot green det viktigaste till att trycka ner sina scorer på touren  och därmed vinna turneringar.

```python
import matplotlib.pyplot as plt
import seaborn as sns

df['Wins'] = df['Wins'].fillna(0)
df['Is_Winner'] = df['Wins'] > 0

tour_snitt = df['Fairway Percentage'].mean()

plt.figure(figsize=(8, 6))

sns.boxplot(data=df, x='Is_Winner', y='Fairway Percentage', color='lightgreen')

plt.axhline(tour_snitt, color='red', linestyle='--', label=f'Tour-snitt ({tour_snitt:.1f}%)')

plt.title('Vinnare har inte högre precision från tee än övriga spelare')
plt.ylabel('Träffade fairways (%)')
plt.xlabel('Har vunnit minst en turnering')
plt.legend()

plt.ylim(50, 70)

plt.show()
```

Jag testade först att göra ett vanligt stapeldiagram, men eftersom det bara visar medelvärdet dolde det helt hur mycket precisionen faktiskt varierar mellan olika spelare. Därför valde jag ett låddiagram (boxplot) för att visa hela spridningen.

Jag har också valt att trunkera y axeln och satt den mellan 50% och 70%. Anledningen är att om y axeln hade börjat på 0 hade det blivit så ihoptryckt att man inte hade sett någon skillnad och inga spelare på denna nivån ligger på 10-20%s nivån över en hel säsong och inte över 90-100%.


Den röda streckade linjen visar tourens verkliga genomsnitt som Python räknat ut till oss med .mean(). Som vi kan se ligger vinnarna och de vinstlösa spelarna nästan helt jämnt runt snittet. Detta visar då att man inte måste vara den med pricksäkra spelaren från tee för att vinna turneringar, så länge man håller sig runt genomsnittet.

#### Avslutning

Min undersökning visar att det finns en tydlig uppoffring, ju längre en spelare slår, desto färre fairways träffas. Trots detta slår spelarna längre för varje år, vilket betyder att man medvetet väljer distans före precision. 

Detta val blir logiskt eftersom precisionen på inspelen har ett starkare samband med en lägre score än vad utslagen har. Mitt sista diagram bekräftar detta då vinnare inte har högre precision från tee än övriga spelare. Slutsatsen är att distansen från tee prioriteras för att få kortare och lättare inspel, vilket leder till lägre score.
